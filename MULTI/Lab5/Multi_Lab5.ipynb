```python
# Função para mostrar vídeo
def display_video(video):
    fig = plt.figure(figsize=(3,3)) 
    mov = []
    for i in range(len(video)): 
        img = plt.imshow(video[i], animated=True)
        plt.axis('off')
        mov.append([img])
    an = animation.ArtistAnimation(fig, mov, interval=50, repeat_delay=1000)
    plt.close()
    return an

#Main (Video completo)
video = imageio.mimread('/content/gdrive/MyDrive/Multimidia/Lab5/bird.avi')  
#video = [resize(frame, (256, 256))[..., :3] for frame in video]    #Ajuste de tamanho (se necessario)
HTML(display_video(video).to_html5_video())
```

```python
# Rertirando 10 Frames
VIDEO = '/content/gdrive/MyDrive/Lab05/videos/bird.avi'
vidcap = cv.VideoCapture(VIDEO)
success,image = vidcap.read()
count = 0
length = int(vidcap.get(cv.CAP_PROP_FRAME_COUNT))
images = []
images2 = []
while success and count < length:
  cv.imwrite("frame%d.jpg" % count, image)       
  success,image = vidcap.read()
  if count < 10:
    images.append(image)
  else:
    images2.append(image)
  print("Frame: ", count)
  count += 1
```

```python
#Os 10 frames retirados(0-9)
HTML(display_video(images).to_html5_video())
```

```python
# O vídeo sem os 10 Frames
HTML(display_video(images2).to_html5_video())
```

##B

```python
# Aplicando filtro de ajuste de cor
imagesWithFilterHSV = []
imagesWithFilterGray = []
for singleImage in images2:
  img = singleImage

  #Converte pra hsv e gray respectivamente
  hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
  gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

  #mostra imagens
  imagesWithFilterHSV.append(hsv)
  imagesWithFilterGray.append(gray)
```

```python
#filro cinza
HTML(display_video(imagesWithFilterGray).to_html5_video())
```

```python
#filro hsv
HTML(display_video(imagesWithFilterHSV).to_html5_video())
```
